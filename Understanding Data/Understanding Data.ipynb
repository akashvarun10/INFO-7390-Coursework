3.  if a company's customer database contains outdated or incorrect contact information, it may result in failed communication attempts and a loss of potential business opportunities.

#### What is Data Cleaning ?
Data Cleaning is the process of identifying and correcting errors, inconsistencies, inaccuracies, and missing values in a dataset to improve its quality and reliability. It is a crucial step in the data preparation phase before using the data for analysis or machine learning.

Data cleaning aims to enhance data quality through error identification and correction, ensuring consistency, completeness, accuracy, and uniformity. It involves handling missing values, removing duplicates, addressing outliers, formatting data, and maintaining documentation. Iterative and requiring domain knowledge, data cleaning is crucial for accurate analysis and reliable machine learning outcomes, ensuring the dataset is suitable for its intended purpose.

### Now Let's Dive into EDA


#### What is Exploratory Data Analysis (EDA) ? 

Exploratory Data Analysis, commonly known as EDA, is an approach in data analysis that employs a variety of techniques to understand and summarize the main characteristics of a dataset, often visually. It's a critical step before formal modeling or hypothesis testing.

Basically , The goal of EDA is to leverage visualization tools, summary tables, and hypothesis testing to:

1. Provide summary level insight into a dataset.

2. Uncover underlying patterns and structures in you data.

3. Identify outliers, missing data, class balance, and other data-related issues.

4. Relate the available data to the business opportunity.



#### Key Techniques in EDA

EDA combines both graphical and quantitative techniques to delve into data:

1. Descriptive Statistics: Involves summarizing main features of data, typically through measures like mean, median, mode, range, and standard deviation. For instance, analyzing average sales per month.
   
2. Data Visualization: Uses graphical representations like histograms, box plots, and scatter plots to understand distributions and relationships within the data.
   
3. Identifying Outliers: Detecting unusual data points that may indicate variability in measurements, experimental errors, or a novelty.

4. Examining Relationships: Exploring correlations or associations between variables using methods like correlation coefficients and cross-tabulation.

5. Group Comparisons: Comparing metrics across different groups. For example, comparing average exam scores across different schools.

#### For Explanation I will be taking Boston Housing Dataset



```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston

# Load Boston Housing dataset
boston = load_boston()
boston_df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
boston_df['PRICE'] = boston.target

# Display the first few rows of the dataset
boston_df.head()
```

#### Descriptive Statistics:
Mean, Median, Mode:
Definition: Measures of central tendency.

#### Useful for: 
Providing a measure of central tendency.
#### Insights:
 Mean represents the average, median is the middle value, and mode is the most frequently occurring value. They give an idea of the dataset's central tendency.


#### Example:
Calculating the mean, median, and mode of house prices.

```python
# Code for Mean House Price
mean_price = boston_df['PRICE'].mean()

# Code for Median House Price
median_price = boston_df['PRICE'].median()

# Code for Mode of House Price
mode_price = boston_df['PRICE'].mode()[0]

# mean_price, median_price, mode_price

print("Mean House Price: ", mean_price)
print("Median House Price: ", median_price)
print("Mode of House Price: ", mode_price)
```

#### Standard Deviation, Variance:
Definition: Measures of dispersion.

#### Useful for: 
Measuring the spread or dispersion of data.
#### Insights: 
High standard deviation/variance indicates more variability in the data. It complements the mean by providing information on how data points deviate from the mean.
#### Example:
Calculating the standard deviation and variance of house prices.

```python
# Code for Standard Deviation of House Prices
std_price = boston_df['PRICE'].std()

# Code for Variance of House Prices
var_price = boston_df['PRICE'].var()

# std_price, var_price


print("Standard Deviation of House Prices: ", std_price)
print("Variance of House Prices: ", var_price)
```

#### Skewness, Kurtosis:
Definition: Measures of the shape of the distribution.

#### Useful for: 
Understanding the shape of the distribution.
#### Insights: 
Skewness measures the asymmetry in the distribution, and kurtosis measures the tail heaviness. They help identify if the data is symmetric or skewed, and if it has heavy or light tails.
#### Example:
Calculating skewness and kurtosis of house prices.

```python
# Code for Skewness of House Prices
skewness_price = boston_df['PRICE'].skew()

# Code for Kurtosis of House Prices
kurtosis_price = boston_df['PRICE'].kurtosis()

skewness_price, kurtosis_price
```

#### Outlier Detection:
Z-Score Analysis:
Definition: Identifies data points that deviate significantly from the mean.

#### Example:
Detecting outliers in house prices using Z-scores.

```python
# Code for Z-Score Analysis
z_scores = (boston_df['PRICE'] - boston_df['PRICE'].mean()) / boston_df['PRICE'].std()
outliers_zscore = boston_df[abs(z_scores) > 3]
print("Number of outliers detected using Z-Score method: ", len(outliers_zscore))
```

#### IQR (Interquartile Range) Method:
Definition: Identifies outliers based on the spread of the data.

#### Example:
Detecting outliers in house prices using the IQR method.

```python
# Code for IQR (Interquartile Range) Method
Q1 = boston_df['PRICE'].quantile(0.25)
Q3 = boston_df['PRICE'].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = boston_df[(boston_df['PRICE'] < Q1 - 1.5 * IQR) | (boston_df['PRICE'] > Q3 + 1.5 * IQR)]
print("Number of outliers detected using IQR method: ", len(outliers_iqr))
```

## Motivation behind EDA
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;John Tukey, while working at Bell Labs, introduced the concept of Exploratory Data Analysis as a technique for effectively using tools of insight to understand a problem before forming hypotheses about the data. The ultimate goal is to gain a deeper understanding of the problem and develop testable hypotheses. Ultimately, the results such as graphs and summary statistics are intended to aid the individual's comprehension and not necessarily to demonstrate a relationship in the data to a general audience.</br>
	&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The ultimate goal of Exploratory Data Analysis (EDA) is to gain valuable insights from the data. To achieve this, EDA typically includes the following steps:
* Identifying and eliminating data outliers
* Detecting trends over time and in different locations
* Finding patterns related to the target variable
* Generating hypotheses and testing them through experiments
* Identifying new sources of data</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The use of various methods to understand and gain perspective on data, both univariate and multivariate, are important in Exploratory Data Analysis (EDA). Examples of such methods include summary statistics (mean, median, min, max, q1, q3), histograms, line charts, box and whisker plots, and scatterplots. EDA is a fundamental aspect of any data analysis that is used to uncover patterns in the data and establish a solid foundation for the overall analysis process. John Tukey, who coined the term EDA, often compared it to detective work. The role of the data analyst or data scientist in EDA is to examine the data from as many angles as possible until a clear narrative emerges, regardless of whether it is ultimately supported by further data.</br>
	&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;When it comes to evaluating the quality of sensor data, Exploratory Data Analysis (EDA) is an effective method. EDA is a way of analyzing data sets to identify their key characteristics, typically through the use of descriptive statistics and metrics or visualizations. These analysis results must be communicated to others such as teammates, stakeholders, and customers. One way to present these results is through the use of metrics like variance, however, using visualizations is often more effective as they are easier to understand, particularly when dealing with multivariate sensor data.</br>
	&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Different types of errors in sensor data can have various underlying causes. For example, outliers may occur when a sensor is operated outside of its specified voltage range, and the level of noise can depend on the applied voltage. The source of the problem in this case is the sensor device. Missing data, on the other hand, can be caused by unreliable data transmission in messaging technologies commonly used in IoT systems, with the root cause being in the architecture or implementation of the distributed system. EDA can assist in identifying these root causes in the overall system. By looking at the same data at different points in time and different stages of the overall system design such as sensor, sensor device, messaging broker component of a distributed system, etc. it is possible to pinpoint the root causes of poor-quality sensor data.
